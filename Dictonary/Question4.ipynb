```python
attendance = {
    "Ali": "Present",
    "Sara": "Absent",
    "Ahmed": "Present",
    "Laiba": "Present"
}
#Print Sara's attendance.
print(f"Sara's Attendance: {attendance.get('Sara')}")

#Add a new student "Hassan" → "Present".
attendance["Hassan"]= "Present"
print(f"\nAddition of new Student: {attendance}")

#Change Ahmed's attendance to "Absent".
attendance["Ahmed"]= "Absent"
print(f"\nUpdated Attendance: {attendance}")

#Remove Ali from the attendance list.
del attendance["Ali"]
print(f"\nUpdated Attendance: {attendance}")
```
