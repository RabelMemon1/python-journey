```python
students = {
    "Ali": 85, 
    "Sara": 92, 
    "Ahmed":78,
    "Laiba": 90
}

print(f"Marks: {students.get('Sara')}")

students['Zain']= 88
print(f"Updated dict:{students} ")
highest_score = max(students, key =students.get)
print(f"Top Student: {highest_score}")
print(f"Highest Marks {students[highest_score]}")
```
