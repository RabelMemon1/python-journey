```python
products = {
    "Laptop": 5,
    "Mouse": 20,
    "Keyboard": 15,
    "Monitor": 7
}

print(f"Stock Available: {products}")
print(f"Stock of mouse: {products.get('Mouse')}")

#Adding a new product
products["Headphones"] = 10
print(f"\nUpdated Stock {products}")

#update Quantity of laptop
products["Laptop"] = 8
print(f"\nUpdated Stock {products}")

#Remove keyboard
del products["Keyboard"]
print(f"\nUpdated Stock Available {products}")
```
