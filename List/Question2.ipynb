```python
#Online Shopping Cart
#An online store stores prices of items in a list.
#cart_prices = [1200, 850, 430, 2999, 650]

#Tasks
#Calculate total bill
#Find the most expensive item
#Remove an item that costs 430
#Add a new item price 1500
```

```python
cart_price = [1200, 850, 430, 2999, 650]
print(f"List: {cart_price}\n")

# Total bill
total_bill = sum(cart_price)
print(f"Total bill: {total_bill}")

#Most Expensive item
exp_item = max(cart_price)
print(f"Expensive Item: {exp_item}")

#Remove an item
remove_item = cart_price.remove(430)
print(f"Remove an item (430): {cart_price}")

#ADD an item
add_item = cart_price.insert(4, 1500)
print(f"Add an item(1500): {cart_price}")
```
