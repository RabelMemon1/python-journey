```python
#Student Grade Analyzer 🎓
#A teacher stores student marks in a list.
#marks = [78, 45, 90, 88, 67, 54, 92]
#Tasks
#Print each student's mark.
#Count how many students passed (marks ≥ 50).
#Count how many students scored above 80.
#Print the average marks of the class.
# Concepts used:
#for loop
#if condition
#counter
#sum() and len()
```

```python
marks = [78, 45, 90, 88, 67, 54, 92]

print(f"Student Marks: {marks}")
passed = 0
above_80 = 0

for mark in marks:
    if mark >= 50:
        passed += 1
    if mark > 80:
        above_80 += 1
print(f"Passed Students: {passed}")
print(f"Highest Scoring Students: {above_80}")

average = sum(marks)/len(marks)
print(f"Average Marks of the class: {average}")
        
        
     
     
```
