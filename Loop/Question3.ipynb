```python
#A grocery store wants to create a simple billing system using Python.
#The system should allow the cashier to enter the prices of items one by one and then display the total bill amount at the end
#Task Requirements
#1.	Ask the user how many items they want to enter.
#2.	Use a loop (either for or while) to input the price of each item.
#3.	Keep a running total of all prices.
#4.	After the loop ends, print the total bill.
#5.	If any item’s price is negative, display a warning message "Invalid price entered!" and skip adding it to the total .	 
```

```python
#rate 5/5

items = int(input("Enter the items YOU want:"))

#LOOP
total = 0
for i in range(items):
    price = int(input(f"Enter the price of item {i+1}:"))
    if price < 0:
        print("Invalid Price!")
        continue


    total = total + price
print(f"Total Bill: {total}")
  
```
