```python
#Even and Odd Counter

#Ask the user to enter 10 numbers.
#Program should:
#count even numbers
#count odd numbers
```

```python

even_count = 0
odd_count = 0
for i in range(1, 11):
    num = int(input(f"Enter {i} number:"))
    if num % 2 == 0:
        even_count = even_count + 1
    else:
        odd_count = odd_count + 1

print(f"Total Even numbers: {even_count}")
print(f"Total Odd numbers: {odd_count}")
  
        
```
