```python
#Largest Number
#Ask the user to enter 5 numbers and find the largest number.
#Example:
#Largest number is: 89
```

```python

max_num = int(input(f"Enter 1 number : "))
for i in range(2,6):
    number = int(input(f"Enter {i} number: "))

    if number > max_num:
        max_num = number
print(f"Maximum Number is : {number}")
```
