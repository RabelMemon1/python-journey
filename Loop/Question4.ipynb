```python
#WHILE LOOP SCENARIOS (2)
️#ATM Withdrawal System 
#A user has balance = 5000.

#Tasks
#Ask the user to enter a withdrawal amount.
#If the amount is less than balance, subtract it.
#Show the remaining balance.
#Keep asking until the user enters 0 (exit).
# Concepts used:
#while loop
#user input
#updating variables
#loop termination
# 4.5 / 5
```

```python

balance = 5000
print("<======ATM Withdrawl System=======>\n")
amount = int(input("Enter a withdrawal Amount:"))

while amount != 0:
    if amount <= balance:
        balance = balance - amount
        print(f"Remaining Balance: {balance}")
    else:
        print("Insufficient Balance!")
    amount = int(input(f"Do You Want to Exit? Enter: 0"))
print(f"\nYou Remaining Account balance is: {balance}")
    
```
