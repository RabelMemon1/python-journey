```python
#Number Guessing Game
#Computer chooses a number between 1–10.
#User keeps guessing until the number is correct.
#Hints:
#Too high
#Too low
#Correct!
```

```python
guess_num = int(input("Enter a guess Number"))

num = int(input("Enter a number (1 to 10)"))

while num != guess_num:
    if num > guess_num:
        print("Too High!")
    elif num < guess_num:
        print("Too Low!")

    num = int(input("Try again"))

print("Congrats you Guessed  the number!")
        
```
