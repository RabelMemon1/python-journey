```python
#Question 1: Simple Calculator
#Take 2 numbers as input from the user and print their 
#sum, difference, product, and division.

#Example Output:
#Enter first number: 10
#Enter second number: 5
#Sum: 15
#Difference: 5
#Product: 50
#Division: 2.0
```

```python
print("<---Simple Calculator --->")
num1 = float(input("Enter the First Number:"))
num2 = float(input("Enter the Second Number:"))

#Addition
sum = num1 + num2
print("\nAddition of two Numbers is: " , sum)
print()

#Difference
difference = num1 - num2
print("Difference of two Numbers is: " , difference)
print()

#Product
multiply = num1 * num2
print("Product of two Numbers is: " , multiply)
print()

#Division

if(num2 == 0):
       print("Division by 0  is not allowed")
else:
    division = num1 / num2   
    print("Division of two Numbers is: " , division)

    
print()


```
