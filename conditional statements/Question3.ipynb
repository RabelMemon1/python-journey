```python
#A cinema has different ticket prices based on age 
#and day of the week:
#Age-based pricing:
#- Children (below 12): Rs. 150
#- Adults (12 to 59): Rs. 300
#- Seniors (60 and above): Rs. 200
#Weekend discount:
#- If it's Saturday or Sunday, give 20% discount
#Take age and day as input, calculate ticket price 
#and print the final amount.
```

```python
#20% discount
discount = 0.20 
age = int(input("Enter your age: "))
day = input("Enter Day of the week: ")

if age >= 60:
    ticket_price = 200
    category ="Senior"
elif age >= 12:
    ticket_price = 300
    category ="Adult"
else: 
    ticket_price = 150
    category ="Child"

print(f"\nBase Price: {ticket_price:.2f}")
print(f"Category: {category}")


if day.lower()== "saturday" or day.lower()== "sunday":
    discounted = ticket_price * discount
    final_price = ticket_price - discounted
    print(f"You have avail 20% discount: {discounted:.2f}")
    print(f"Final bill: {final_price:.2f}")
else:
     print(f"Total bill is: {ticket_price:.2f}")


  

    
    

```
