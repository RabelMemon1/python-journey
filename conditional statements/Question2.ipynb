```python
#Ask the user to enter:
#Customer Name
#Total Purchase Amount
#Apply the following discount rules:
#If the total amount is 5000 or more, give a 20% discount.
#If the total amount is between 3000 and 4999, give a 10% discount.
#If the total amount is between 1000 and 2999, give a 5% discount.
#If the total amount is less than 1000, no discount is applied.
#After calculating the discount:
#Display the discount amount.
#Display the final amount to be paid.
#At the end, print:
```

```python
discount1 = 0.20  # discount 20%
discount2 = 0.10 # discount 10%
discount3 = 0.05 # discount 5%
discounted = 0
customer_name = input("What is your Name: ")
purchase_amount = float(input("Total Amount of Purchase: "))


print("\nCustomer Name: " , customer_name)
print("Purchase Amount: " , purchase_amount)

if purchase_amount >= 5000:
    discounted = purchase_amount * discount1
    print("You recieved 20% discount: " , discounted)

elif purchase_amount >= 3000:
    discounted = purchase_amount * discount2
    print("You recieved 10% discount: " , discounted)

elif purchase_amount >= 1000:
    discounted = purchase_amount * discount3
    print("You received 5% discount: " , discounted)
else:
    print("No discounts applied ")

total_bill = purchase_amount - discounted
print("Total bill: " , total_bill)


    
```
