```python
#Take a student's marks as input and print their grade 
#based on the following criteria:

#- 90 and above: Grade A
#- 80 to 89: Grade B
#- 70 to 79: Grade C
#- 60 to 69: Grade D
#- Below 60: Grade F

#Example Output:
#Enter your marks: 85
#Your grade is: B
```

```python
print("GRADE CALCULATION")
marks = int(input("Enter your marks in Mathematics course:" ))
if(marks >= 90):
    print("A")
elif(marks <= 89 and marks >= 80):
    print("B")
elif(marks <= 79 and marks >= 70):
    print("C")
elif(marks <= 69 and marks >= 60):
    print("D")
else: 
    print("F")
```
