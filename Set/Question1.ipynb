```python
#	You are a web data analyst working for a news platform. Each day, the system records the user IDs of visitors.
#Some users visit every day, while others appear only once. Your task is to analyze visitor behavior using Python sets.
#1.	Store visitors’ user IDs for three days using sets, take 3 days data as input from the user.
#2.	Perform the following analyses:
#	Find total unique visitors across all days (union)
#	Find users common to all days (intersection)
#	Find users who visited only on Day 1 (difference)
#	Find users who visited on exactly one of Day 1 or Day 2 (symmetric_difference) 
#Check if Day 1 and Day 3 have no common visitors (isdisjoint)
#1. Display results clearly and label each output.

#day1_visitors = {"user1","user2", "user3","user4" }
#day2_visitors = {"user3","user4", "user5","user6" }
#day3_visitors = {"user7","user8", "user5"}
```

```python
day1_visitors = {"user1","user2", "user3","user4" }
day2_visitors = {"user3","user4", "user5","user6" }
day3_visitors = {"user7","user8", "user5"}

##	Find total unique visitors across all days (union)
unique_visitor = day1_visitors.union(day2_visitors, day3_visitors)
print(f"Total unique Visitore: {unique_visitor}")

#	Find users common to all days (intersection)
common_visitor = day1_visitors.intersection(day2_visitors, day3_visitors)
print(f"Total Common Visitor: {common_visitor}")

#	Find users who visited only on Day 1 (difference)

first_day_visitor = day1_visitors.difference(day2_visitors, day3_visitors)
print(f"First Day Visitors: {first_day_visitor}")

#	Find users who visited on exactly one of Day 1 or Day 2 (symmetric_difference)
one_time_visitor = day1_visitors.symmetric_difference(day2_visitors)
print(f"One Visit in 2 days: {one_time_visitor}")

#Check if Day 1 and Day 3 have no common visitors (isdisjoint)
check_common_visitor = day1_visitors.isdisjoint(day3_visitors)
print(f"Checking Common visitors on day 1 and 3: {check_common_visitor}")
```
